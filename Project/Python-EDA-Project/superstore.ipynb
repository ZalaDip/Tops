# Visualizing Data

```python
plt.figure(figsize=(15,6))
sns.countplot(x='Ship_Mode', data=ss)
plt.xticks(rotation=45)
plt.title("Number of Order by Ship Mode")
plt.grid()
plt.show()
```

--> From this graph we can conclude that:

1. Standard Class is the most preferred shipping mode, accounting for the highest number of orders—nearly 6,000.

2. Second Class and First Class follow, with significantly fewer orders, around 2,000 and 1,500 respectively.

3. Same Day shipping is the least used mode, indicating limited demand or availability for urgent deliveries.

```python
plt.figure(figsize=(20,10))
sns.barplot(x=ss["Country"].value_counts()[:20].index, y=ss["Country"].value_counts()[:20].values)
plt.title("Country locates Customers")
plt.xlabel('Countries')
plt.ylabel("Count")
plt.xticks(rotation = 50)
plt.show()
```

1. All customers are located in the United States, as it is the only country represented in the data.

2. The total customer count reaches approximately 10,000, indicating a U.S.-centric customer base.

```python
group_data = ss.groupby("Ship_Mode")["Sales"].sum().reset_index()
format_data = group_data.pivot_table(columns="Ship_Mode", values="Sales", aggfunc='sum')
format_data
```

```python
format_data.info()
```

```python
sum_data = format_data.sum()

labels = sum_data.index.tolist()
values = sum_data.values.tolist()

plt.figure(figsize=(10,8))
plt.pie(values, labels=labels, autopct="%1.2f%%",explode=[0,0.2,0,0])
plt.title("Types of Product Category Sales")
plt.legend()
plt.axis('equal')
plt.show()
```

```python
59.12+19.99
```

--> Upto 79% of the Sales are result of the Standard Class and Second Class Ship Modes.

# Key Insights

### 1. Order Volume by Ship Mode

--> Standard Class dominates shipping preferences with nearly 6,000 orders, suggesting it is the most cost-effective or widely offered option.

--> Second Class and First Class follow with moderate usage.

--> Same Day shipping is the least used, indicating either limited availability or cost sensitivity.

### 2. Customer Location Distribution

--> All customers are based in the United States, confirming that the business operates within a single-country market.

--> A total of over 10,000 customer records emphasize a broad domestic reach.

### 3. Customer Segment Distribution

--> The Consumer segment holds the highest count, followed by Corporate and Home Office.

--> This indicates the store’s primary target market is individual consumers, though B2B clients are also significant.

### 4. Sales Distribution

--> Sales are heavily concentrated among a few values, with a small number of transactions contributing a large share of revenue.

--> A pie chart indicates that the top 3 sales values make up a considerable percentage of total transactions.

### 5. Quantity Ordered

--> The most common ordered quantities range between 2 and 3 items.

--> A pie chart reveals that the top 4 most frequently ordered quantities account for the majority of sales volume.

### 6. Top Customers

--> A small group of high-frequency customers contributes significantly to overall order volume.

--> The top 4 customers alone form a major chunk of the total customer order count.

### 7. Sales by Ship Mode

--> When grouped by shipping mode, Standard Class not only has the highest order volume but also the highest total sales revenue.

--> This reflects its importance in both order frequency and financial performance.
