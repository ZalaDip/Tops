```python
group_data = asd.groupby("PRODUCTLINE")["SALES"].sum().reset_index()
format_data = group_data.pivot_table(columns="PRODUCTLINE", values="SALES", aggfunc='sum')
format_data
```

```python
format_data.info()
```

```python
plt.style.available
```

```python
plt.style.use("seaborn-v0_8-colorblind")
```

```python
sum_data = format_data.sum()

labels = sum_data.index.tolist()
values = sum_data.values.tolist()

plt.figure(figsize=(10,8))
plt.pie(values, labels=labels, autopct="%1.2f%%",explode=[0,0,0,0,0.2,0,0])
plt.title("Types of Product Category Sales")
plt.legend()
plt.axis('equal')
plt.show()
```

# INSIGHTS

## Data Overview

#### --> The dataset contains various features like Price, Mileage, Year, Make, and Model of cars

#### --> Initial inspection includes:

1. Checking for null values.
2. Basic statistical summary using describe()
3. Unique value counts for categorical variables like Make and Model.

## Data Cleaning & Preparation

#### --> Null values are handled appropriately - either removed or filled.

#### --> Outliers and anomalies in columns like Price and Mileage are examined through visualizations(Ex., Boxplots).

## Exploratory Data Analysis (EDA)

#### --> Price is positively correlated with Year(Newer cars are more expensive), and negatively correlated with Mileage

#### --> Visualizations:

1. Scatter plots of Price vs. Year
2. Bar plots of average price by Make and Model.
3. Distribution plot show a right price distribution(more lower-priced cars).

## Modeling & Prediction

#### --> Multiple regression models(Ex., Linear Regression) are used to predict car prices.

#### --> Feature importance or coefficients are analyzed, highlighting key predictors such as:

1. Year of manufacture
2. Mileage
3. Make and Model

## Key Insights

#### 1. Newer cars and those with lower mileage are priced significanly higher

#### 2. Certain brands(Toyota, Honda, etc.) trend to have higher resale value.

#### 3. Mileage has a strong negative impact on price - suggesting it's crucial variable in price prediction.
